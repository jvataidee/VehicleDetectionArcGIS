```python
from arcgis.learn import prepare_data, FasterRCNN
```

```python
filepath = r'C:\Users\jv_at\tcc\Arcgis\Amostra_total'
```

```python
data = prepare_data(filepath, 
                    batch_size= 1, 
                    dataset_type="PASCAL_VOC_rectangles", 
                    chip_size = 540)
```

```python
model = FasterRCNN(data, backbone = "resnet152")

lr = model.lr_find()
print("Learning Rate:",lr)

model.fit(20, lr =lr) 

model.average_precision_score()
```

```python
model.learn.recorder.plot_losses()
```

```python
model.show_results()
```

```python
model.save('model_finale')
```
